```python
import pandas as pd
data=pd.read_csv('enhanced_zomato_dataset_clean.csv')
```

```python
# check for duplicate records where Restraunt name and Item_Name are same and only show their index and the records
data[data.duplicated(subset=['Restaurant_Name', 'Item_Name'], keep=False)].sort_values(by=['Restaurant_Name', 'Item_Name'])
```

```python
# keep only the first occurence of the duplicate records and drop the rest
data.drop_duplicates(subset=['Restaurant_Name', 'Item_Name'], keep='first', inplace=True)
data
```

```python
# check for duplicate records where Restraunt name and Item_Name are same and only show their index and the records
data[data.duplicated(subset=['Restaurant_Name', 'Item_Name'], keep=False)].sort_values(by=['Restaurant_Name', 'Item_Name'])
```

```python
from sqlalchemy import create_engine
# Database connection parameters
DB_USER = "root"
DB_PASSWORD = "nitya"
DB_HOST = "localhost"       # or IP address
DB_PORT = 3306              # default MySQL port
DB_NAME = "zepto"


engine = create_engine(
        f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}",
        echo=True,           # Set to True to log SQL queries
        pool_pre_ping=True,  # Helps avoid stale connections
        pool_recycle=3600    # Recycle connections every hour
    )
```

```python
# send data to mysql database
data.to_sql(name='Zepto_Data', con=engine, if_exists='append', index=False)
```

```python
Unique_cuisines = pd.read_sql("SELECT DISTINCT cuisine FROM Zepto_Data", con=engine)
```
