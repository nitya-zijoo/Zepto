```python
from sqlalchemy import create_engine
# Database connection parameters
DB_USER = "root"
DB_PASSWORD = "nitya"
DB_HOST = "localhost"       # or IP address
DB_PORT = 3306              # default MySQL port
DB_NAME = "zepto"


engine = create_engine(
        f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}",
        echo=True,           # Set to True to log SQL queries
        pool_pre_ping=True,  # Helps avoid stale connections
        pool_recycle=3600    # Recycle connections every hour
    )
```

```python
import pandas as pd
data= pd.read_sql("SELECT * FROM Zepto_Data", con=engine)
```

```python
# fetch unique no of restraunts from the data using sql query and store it in a variable
unique_restaurants = pd.read_sql("SELECT cuisine, count(Restaurant_Name) as restaurant_count FROM Zepto_Data  group by cuisine having count(Restaurant_Name) < 10", con=engine)
```
